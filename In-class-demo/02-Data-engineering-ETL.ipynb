```python
import pandas as pd
```

```python
data = pd.read_excel("datasets/01-attendantsheet-sample.xlsx")
```

```python
data.head(20)
```

```python
data.iloc[6:22, :34]
```

```python
start_row = 0
end_row = 0
end_column = 0
```

```python
data.shape
```

```python
total_rows, total_columns = data.shape
total_rows, total_columns
```

```python
for row in range(total_rows):
    sequence_number = data.iloc[row, 0]
    if "date" in str(sequence_number).lower():
        start_row = row
        print(f"Found Start Row at {row}")

    if "legend" in str(sequence_number).lower():
        end_row = row
        print(f"Found End Row at {row}")
```

```python
data.iloc[start_row:end_row, :end_column]
```

```python
filtered_data = data.iloc[start_row:end_row]
filtered_data.head()
```

```python
# total_filtered_column = filtered_data.shape[1]
# total_filtered_column
```

```python
for column in range(total_columns):
    colummn_header = filtered_data.iloc[0, column]
    if str(colummn_header).lower() == 'total':
        print(f"Found End Column: {column}")
        end_column = column
```

```python
attendant_dataset = filtered_data.iloc[:, :end_column]
attendant_dataset
```
