
10. Find The Email of the person with the following IP Address: 132.207.160.22

```python
IP_ADDRESS = '132.207.160.22'

selected_address = df['Email'][df['IP Address'] == IP_ADDRESS]
# Initially, it is a Series

email = selected_address.iloc[0]
# So, if we just want the value (string), just iloc it!

print(f"The Email of the dude with IP Address {IP_ADDRESS} is {email}")
```


11. How many People have Mastercard as their Credit Card Provider and made a purchase above 50?

```python
MC_and_above_50 = df[(df['CC Provider'] == 'Mastercard') & (df['Purchase Price'] > 50)]
# LET'S GOOOOO, Using the Bitwise AND Operator (&) with Boolean Indexing
# For example: [() & ()]

print(f"There are {len(MC_and_above_50)} dudes who have Mastercard and made a purchse above 50")
```


12. Find the email of the person with the following Credit Card Number: 4664825258997302


13. How many people purchase during the AM and how many people purchase during PM?


14. How many people have a credit card that expires in 2020?


15. What are the top 5 most popular email providers (e.g. [gmail.com](http://gmail.com/), [yahoo.com](http://yahoo.com/), etc...)

```python
# 1. extract into a providers column using regex
# 2. groupby the providers
# LEt's goooo

# df['Provider'] = df['Email'].str.extract(pat=r'@([\w.-\.]+)')

# The "Greedy" Regex
df['Provider'] = df['Email'].str.extract(r'@([^ ]+)')
# [^ ] means "any character except a space"
# + means "one or more"

providers_grouped = df.groupby('Provider').size().sort_values(ascending=False).head()
providers_grouped
```
