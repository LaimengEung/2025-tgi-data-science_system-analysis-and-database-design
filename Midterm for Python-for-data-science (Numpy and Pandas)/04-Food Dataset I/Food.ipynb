<b>Exercise 5. Display the <i>"food"</i> which has the highest number of orders?

```python
# Only returns the food name
# order_list['food'].mode()

# Display it with the number of occurences too
food_count = order_list['food'].value_counts()
highest_order = food_count.head(1)

print("the 'food' which has the highest number of orders is ", highest_order.index[0])
print("It was ordered", highest_order.iloc[0], "times")
```

<b>Exercise 5. How many items were ordered for the most-ordered item?

```python
print(highest_order.index[0], "was ordered", highest_order.iloc[0], "times")
```

<b>Exercise 6. Display the <i>"description"</i> which has the highest number of orders?</b>

```python
highest_quantity = order_list['quantity'].max()
desc_of_highest_order = order_list[order_list['quantity'] == highest_quantity]['description'].values[0]

# order_list[order_list['quantity'] == highest_quantity]['description'] returns a Series; .values[0] gets the first value
type(desc_of_highest_order)

print(highest_quantity)
print(desc_of_highest_order)
```

```python
# ORRRRRR, using idxmax(): find the index label of the maximum value
idx = order_list['quantity'].idxmax() # find the index label

desc_of_highest_order = order_list.loc[index:=idx, col:='description']
                                        # index:= and col:=; or not, it's fine
# or: using .iloc
# desc_of_highest_order = order_list.iloc[idx]['description']
highest_quantity = order_list.loc[idx, 'quantity']

print(f"Description: {desc_of_highest_order}")
print(f"Quantity: {highest_quantity}")
```

<b>Exercise 7. How many foods ordered in total?</b>

```python
total_food_ordered = order_list['quantity'].sum()

print(f"Total quantity ordered: {total_food_ordered}")
```

<b>Exercise 8. Calculate the total income for the period in the dataset?</b>

```python
# total_income = order_list['unit_price'].multiply(order_list['quantity']).sum()
# Cannot, because elements of order_list['unit_price'] are strings

# 1. Convert all elements of order_list['unit_price'] to float and slice $ out
    # Or convert it to string then replace '$' with ''
# 2. Multiply it with quantity
# 3. Sum()

prices = order_list['unit_price'].str.replace('$', '').astype(float)

total_income = (prices * order_list['quantity']).sum()

print(f"Total Income: ${total_income}")
```

<b>Exercise 9. Find the total number of orders.</b>

```python
total_num_of_order_by_row = len(order_list)
total_num_of_order_by_id = order_list['id'].nunique()

print(f"There are {total_num_of_order_by_row} orders by Orders?")
print(f'There are {total_num_of_order_by_id} orders by Invoice?')
```

<b>Exercise 10. Find the average income for each order?</b>

```python
avr_income_each_order = (prices * order_list['quantity']).mean()

print(f"Average Income In Each Order: ${avr_income_each_order:.2f}")
```

<b>Exercise 11. How many different foods are ordered?</b>

```python
unique_foods = order_list['food'].nunique()

print(f"There are {unique_foods} different foods ordered")
```
