<b>Exercise 2. Create the 3 DataFrames from the dictionaries.</b>

```python
data_1_dict = {'student_id': range(1,11),
               'first_name': ['Mony', 'Nana', 'Ray', 'Alice', 'Ratanak', 'Nara', 'Ratana', 'Roth', 'Rith', 'Johnny'],
               'last_name': ['Sopheak', 'Tanaka', 'Allen', 'May', 'Sam', 'Kun', 'Vann', 'Sovann', 'Harrison', 'Cole']}

df_1 = pd.DataFrame(data_1_dict)

df_1
```

```python
data_2_dict = dict([('student_id', range(11, 16)),
                   ('first_name', ['John', 'Frank', 'James', 'Tom', 'Jennifer']),
                    ('last_name', ['Bond', 'Little', 'Collar', 'Brune', 'Stone'])])

df_2 = pd.DataFrame(data_2_dict)

df_2
```

```python
data_3_dict = dict(subject_id=range(1,11),
                   subject_name=['Math', 'Geography', 'History', 'Literature', 'Research', 'IT', 'Moral', 'Art', 'Earth Science', 'Physics'])

df_3 = pd.DataFrame(data_3_dict)

df_3
```

```python
data_4_dict = {'student_id': range(1, 11),
               'subject_id': [1, 3, 7, 8, 5, 4, 2, 7, 1, 6],
               'test_score': [55, 75, 65, 93, 76, 80, 77, 81, 63, 89]}

df_4 = pd.DataFrame(data_4_dict)

df_4
```

<b>Exercise 3. Concat data1 and data2 Dataframes placing the data2 under data1.</b>

```python
data2_under_data1 = pd.concat([df_1, df_2])

data2_under_data1
```

<b>Exercise 4. Concat data1 and data2 Dataframes placing the data2 on the right side of data1.</b>

```python
data2_right_of_data1 = pd.concat([df_1, df_2], axis=1)
# Normally axis=0 (concat by rows), axis=1 concat by columns (to the right)

data2_right_of_data1
```

<b>Exercise 5. Merge the result in <i>Exercise 3</i> with data4 and display the result.</b>

```python
# ex3_merge_df_4 = pd.merge(data2_under_data1, df_4)

ex3_merge_df_4 = pd.merge(data2_under_data1, df_4, on='student_id')
    # on= The column to merge on (student_id, which exists on both)
    # will only show id from 1-10, since df_4 only have 1-10
ex3_merge_df_4
```

<b>BONUS: Make a Complete Table, with the Student Information, as well as the subject and the test score</b>

```python
subject_table = pd.merge(df_4, df_3, on='subject_id')

# ----- DisRegard this Comment :D (I wrongfully named subject_id as student_id for df_3)[It's changed now!] ------
# subject_table[['student_id_x', 'subject_id', 'subject_name', 'test_score']] # to show only some columns name
# merge on subject_id from df_4 with student_id from df_3 to get the subject names

# on: The column names to join on. 
# left_on and right_on: Columns in the left and right DataFrames, respectively, 
#                       to use as keys when their names are different.

student_table = pd.concat([df_1, df_2])

complete_table = pd.merge(student_table, subject_table, on='student_id')

complete_table
```

<b>Exercise 6. How many points does Johnny Cole get in the test? Display the information, score and subjects of Johnny Cole.</b>

```python
j_cole_info = ex3_merge_df_4[(ex3_merge_df_4['first_name'] == 'Johnny') & (ex3_merge_df_4['last_name'] == 'Cole')]
# Use & (Bitwise) in pandas; it doesn't support 'and' like in normal Python
# Must () & () as well [() between conditions]

j_cole_info
```
